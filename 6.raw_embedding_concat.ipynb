```python
# Raw Concatenation of Embeddings
import numpy as np
from pathlib import Path

data_dir = Path("data")

# 1) Load each embedding matrix
gpt2_emb    = np.load(data_dir / "gpt2_embeddings.npy")    # (N, D1)
sbert_emb   = np.load(data_dir / "sbert_embeddings.npy")   # (N, D2)
graph_emb   = np.load(data_dir / "graph_embeddings.npy")   # (N, 128)
meta_emb    = np.load(data_dir / "meta_embeddings.npy")    # (N, M)

# 2) load  ID mapping to keep row→champion alignment
ids         = np.load(data_dir / "champion_ids.npy")       # (N,)

# 3) Concatenate along feature‐axis
combined_raw = np.concatenate([gpt2_emb, sbert_emb, graph_emb, meta_emb], axis=1)
print("Combined raw embeddings shape:", combined_raw.shape)

# 4) Save out for downstream use
np.save(data_dir / "combined_raw_embeddings.npy", combined_raw)
np.save(data_dir / "combined_ids.npy", ids)
```
